Challenge: Student Marks Analyzer

Create a Python program that allows a user to input student names along with their marks and then calculates useful statistics.
Your program should:
1. Let the user input multiple students with their marks (name + integer score).
2. After input is complete, display:
   - Average marks
   - Highest marks and student(s) who scored it
   - Lowest marks and student(s) who scored it
   - Total number of students

Bonus:
- Allow the user to enter all data first, then view the report
- Format output clearly in a report-style layout
- Prevent duplicate student names


```python
class Student:
    def __init__(self,name,marks):
        self.name = name
        self.marks = marks


class StudentAnalyzer:
    def __init__(self):
        self.students = []

    def add_students(self,name,marks):
        for s in self.students:
            if s.name ==name:
                print("Student already exist!")
                return

        self.students.append(Student(name,marks))
        print(f"{name} added successfully!")

    def show_report(self):
        if not self.students:
            print("No data available")
            return

        total = len(self.students)
        marks_list = [s.marks for s in self.students]

        avg = sum(marks_list) / total
        highest = max(marks_list)
        lowest = min(marks_list)

        highest_students = [s.name for s in self.students if s.marks == highest]
        lowest_students = [s.name for s in self.students if s.marks == lowest]

        print("\n------Student Report------")
        print(f"Total Students: {total}")
        print(f"Average Marks: {avg:.2f}")

        print(f"\nHighest Marks: {highest}")
        print("Topper(s):",",".join(highest_students))

        print(f"\nLowest Marks: {lowest}")
        print("Lowest Scorer(s):", ", ".join(lowest_students))

analyzer = StudentAnalyzer()

while True:
    print("\n1. Add Students")
    print("2. Show Report")
    print("3. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        name = input("Enter name: ")
        try:
            marks = int(input("Enter marks: "))
        except ValueError:
            print("Invalid marks!")
            continue

        analyzer.add_students(name, marks)

    elif choice == "2":
        analyzer.show_report()

    elif choice == "3":
        print("Exiting...")
        break

    else:
        print("Invalid choice!")

        
```
