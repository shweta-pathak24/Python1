 Challenge:  Personal Movie Tracker with JSON

Create a Python CLI tool that lets users maintain their own personal movie database, like a mini IMDb.

Your program should:
1. Store all movie data in a `movies.json` file.
2. Each movie should have:
   - Title
   - Genre
   - Rating (out of 10)
3. Allow the user to:
   - Add a movie
   - View all movies
   - Search movies by title or genre
   - Exit the app

Bonus:
- Prevent duplicate titles from being added
- Format output in a clean table
- Use JSON for reading/writing structured data



```python
import json
import os

file_name = "movies.json"

class Movie:
    def __init__(self,title, genre, rating):
        self.title = title
        self.genre = genre
        self.rating = rating

    def to_dict(self):
        return{
            "title": self.title,
            "genre": self.genre,
            "rating": self.rating

        }

class MovieTracker:
    def __init__(self):
        self.movies = self.load_movies()

    def load_movies(self):
        if not os.path.exists(file_name):
            return[]
        with open (file_name, "r") as file:
            data = json.load(file)
            return [Movie(**movie) for movie in data]

    def save_movies(self):
        with open(file_name, "w") as file:
            json.dump([movie.to_dict() for movie in self.movies], file,indent=4)

    def add_movie(self):
        title = input("Enter title: ").strip()

        for movie in self.movies:
            if movie.title.lower() == title.lower():
                print("Movie already exists! ")
                return

        genre = input("Enter genre: ").strip()

        try:
            rating = float(input("Enter rating (out of 10): "))
        except ValueError:
            print("Invalid rating!")
            return

        new_movie = Movie(title, genre, rating)
        self.movies.append(new_movie)
        self.save_movies()

        print("Movie added!")

    def view_movies(self):
        if not self.movies:
            print("No movies found!")
            return

        print("\n Movie found!")
        print("-" * 50)
        print(f"{'Title':20} {'Genre': 15} {'Rating': >5}")
        print("-" * 50)

        for movie in self.movies:
            print(f"{movie.title:20} {movie.genre:15} {movie.rating:>5}")

        print("-" * 50)


    def search_movies(self):
        keyword = input("Search: ").lower()

        results = [
            movie for movie in self.movies
            if keyword in movie.title.lower() or keyword in movie.genre.lower()

        ]
        if not results:
            print("No results found!")

        print("\n Results")
        print("-" * 50)
        print(f"{'Title': 20} {'Genre':15} {'Rating':>5}")
        print("-" * 50)

        for movie in results:
            print(f"{movie.title:20} {movie.genre:15} {movie.rating:>5}")
        print("-" * 50)


    def main():
            tracker = MovieTracker()

            while True:
                print("\n Movie Tracker")
                print("1. Add Movie")
                print("2.View Movies")
                print("3.Search Movies")
                print("4. Exit")

                choice = input("Enter choice: ")

                if choice == "1":
                    tracker.add_movie()

                elif choice == "2":
                    tracker.view_movies()

                elif choice == "3":
                    tracker.search_movies()

                elif choice == "4":
                    print("Goodbye")
                    break
                else:
                    print("Invalid choice")


    if __name__ == "__main__":
            main()
    


       





    
    
           
```
