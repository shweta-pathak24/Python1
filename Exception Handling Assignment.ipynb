```python
try:
    with open("shweta.txt","r") as file:
        print(file.read())

except FileNotFoundError:
    print("File not found!")
```

4. Write a Python program that prompts the user to input two numbers and raises a TypeError exception if the inputs are not numerical.


```python
try:
    num1 = input("Enter number 1:")
    num2 = input("Enter number 2: ")


    if not num1.isnumeric() and not num2.isnumeric():
        raise TypeError("Invalid number!")
    
    num3 = float(num1)
    num4 = float(num2)

    print(f"Your input:{num3},{num4}")
    
except TypeError as e:

    print(f"Error:{e}")
    
```

5. Write a Python program that opens a file and handles a PermissionError exception if there is a permission issue.


```python
try:
    with open("data.txt","r") as file:
        line = file.read()
        print(line)
        
except PermissionError:
    print("You don't get a permission to access this file")

except FileNotFoundError:
    print("File not found")
```

6. Write a Python program that executes an operation on a list and handles an IndexError exception if the index is out of range.


```python
try:
    lst = [100,200,300,400,500,600]

    index = int(input("Enter the index:"))

    print("Element at index:",lst[index])

except IndexError:
    print("Error: Index is out of range")
```
